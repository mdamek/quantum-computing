```python
%matplotlib inline
# Importing standard Qiskit libraries and configuring account
from qiskit import QuantumCircuit, execute, Aer, IBMQ
from qiskit.compiler import transpile, assemble
from qiskit.tools.jupyter import *
from qiskit.visualization import *
# Loading your IBM Q account(s)
provider = IBMQ.load_account()
```

```python
provider = IBMQ.get_provider(group='open')
backendSimulator = provider.get_backend('ibmq_qasm_simulator')
backendReal = provider.get_backend('ibmq_rome')
```

```python
circ1 = QuantumCircuit(2,1)
circ1.x(1)
circ1.h(0)
circ1.h(1)
circ1.cnot(0,1) #bramka obliczeniowa Uf1, należy do grupy funkcji zmiennych
circ1.h(0)
circ1.h(1)
circ1.measure(0,0)
circ1.draw()
```

```python
job = execute(circ1, backendSimulator)
result = job.result()
counts = result.get_counts(circ1)
plot_histogram(counts)
#Wynik to 1: funkcja zmienna
```

```python
job = execute(circ1, backendReal)
result = job.result()
counts = result.get_counts(circ1)
plot_histogram(counts)
```

Problem Deutcha polega na znalezieniu klasy funkcji, czyli tego czy jest zmienna czy stała. Polega również na określeniu ile razy musimy sprawdzić wartości funkcji, aby się tego dowiedzieć. Funkcje te są następujących wartości:
f(0)f(1)
f0 0 0 stała
f1 0 1 zmienna

Wyniki z komputera kwantowego potwierdzają wyniki symulacyjne. Oczywiście nie jest to w 100% z powodu błędów jakie występują przy realnych komputerach kwantowych, jednak wskazania są zdecydowane na oczekiwane wartości.
